```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
```

```python
revenue_data = pd.read_csv('../Data/Raw/Tickets.csv')
# add rolling revenue
revenue_data['Revenue'] = np.where(revenue_data['net_weight']>0, revenue_data['net_weight']*revenue_data['rate'], 
                                   revenue_data['hours_worked']*revenue_data['rate'])
```

```python
revenue_data['date'] = pd.to_datetime(revenue_data['date'])
revenue_data.set_index('date', inplace=True)
```

# Year To Date Revenue

```python
# If you want to change the start date, modify the start_date variable. ie start_date='2022-01-10'
start_date = '2022-01-01'
```

```python
# DO NOT MODIFY
end_date = dt.date.today()
ytd_revenue = revenue_data['Revenue'] 
```

```python
# DO NOT MODIFY
sns.set(rc={'figure.figsize':(11, 8)})
sns.lineplot(x=ytd_revenue.index, y=ytd_revenue.values.cumsum())
```

# Revenue by Time Period

 Modify the start and end date.
 
Be sure to use ' or " at the start and end of period


The format is YYYY-MM-DD

```python
# Modify this cell
start_date = '2022-01-01'
end_date = '2022-01-03'
```

```python
# Do NOT MODIFY
time_period_data = revenue_data.loc[start_date:end_date]
time_period_data = time_period_data.sort_values(by='date')
time_period_revenue = time_period_data['Revenue']
```

```python
# This runs the plot
time_period_revenue.cumsum().plot()
```
